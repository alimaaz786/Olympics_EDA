Top 10 countries participating in olympics

```python
top_10=df["Team"].value_counts().sort_values(ascending=False).head(10)
plt.style.use("ggplot")
plt.title("Top 10 countries particiapted in olympics",size=25,color="Green")
top_10.plot(kind="barh")
```

```python
gold_medals=df.query('Medal=="Gold"')
top_10_c=gold_medals.Region.value_counts().head(10)
fig=plt.figure(figsize=(18,8))
top_10_c.plot(kind="bar")
plt.title("Top 10 countries with most gold medals",size=25,color="green")
plt.xlabel("gold medals")
```

```python
silver_medals=df.query('Medal=="Silver"')
top_10_s=silver_medals.Region.value_counts().head(10)
fig=plt.figure(figsize=(12,8))
top_10_s.plot(kind="bar")
plt.title("Top 10 countries with silver medals",size=25,color="green")
plt.xlabel("silver medals")
```

so we can observe from this that the countries who have won gold medals have also won the silver medals most

```python
#Gender wise participation
gender=df["Sex"].value_counts()
gender
```

```python
gender.plot(kind="pie",autopct = '%1.1f%%' , shadow = True)
plt.title("total gender")
```

```python
women=df.query('Sex=="F" & Season=="Summer"')
women.head(10)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x="Year",data=women)
```

```python
men_over_time=df.query('Sex=="M" & Season=="Summer"')
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
p=men_over_time.groupby("Year")["Sex"].value_counts()
p.loc[:,"M"].plot()
plt.title("men participation over time",size=20,color="green")
```

```python
women_over_time=df.query('Sex=="F" & Season=="Summer"')
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
w=women_over_time.groupby("Year")["Sex"].value_counts()
w.loc[:,"F"].plot()
plt.title("women participation over time",size=20,color="Green")
```

conclusion--> Over the year the particpation number of men and women have been increased but the increasing rate of women is much more faster then men
